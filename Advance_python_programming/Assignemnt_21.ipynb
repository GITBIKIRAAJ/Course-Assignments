Q2. **Is there a distinction between true random numbers and pseudo-random numbers, if there is one? Why are the latter considered “good enough”?**

Ans: True Random Numbers are true physical values while pseudo-random numbers are generated internally by a program . Latter is considered good enough as they have sort of uniform distribution. Pseudo random numbers are generated by computers using an algorithm (there are many) and a seed (either chosen by the user or, sometimes, clock time). 

They are not strictly random as, if you start with the same seed and same algorithm. A truly random number, on the other hand, is completely unpredictable. There is no way to generate a series of truly random numbers via computer; the usual ways are flipping coins, throwing darts, rolling dice and other physical processes.

Q4. **Provide a real-life example of a normal distribution ?**

**Ans**:A normal distribution is a probability distribution that is symmetrical around the mean and has a bell-shaped curve. Normal distributions are often used to model real-world phenomena that are continuous and follow a predictable pattern.

One example of a real-life situation that can be modeled using a normal distribution is the height of people in a population. The heights of people in a population tend to follow a normal distribution, with the majority of people having heights that are close to the average and a smaller number of people having heights that are significantly above or below the average.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values from a normal distribution
# with a mean of 175 cm and a standard deviation of 5 cm
heights = np.random.normal(175, 5, 1000)

# Plot the distribution of heights
plt.hist(heights, bins=30)
plt.xlabel('Height (cm)', fontsize= 15)
plt.ylabel('Frequency', fontsize= 15)
plt.show()
```

Q5. **In the short term, how can you expect a probability distribution to behave? What do you think will happen as the number of trials grows ?**

**Ans**: A probability distribution is a function that describes the likelihood of different outcomes in a random event. Depending on the type of probability distribution, you can expect it to behave in different ways.

For example, if you are working with a uniform probability distribution, you can expect the probability of each outcome to be the same. In this case, as the number of trials grows, the relative frequency of each outcome should approach the probability of that outcome.

On the other hand, if you are working with a skewed probability distribution, you can expect some outcomes to be more likely than others. In this case, as the number of trials grows, the relative frequency of the more likely outcomes should approach their probabilities more closely, while the relative frequency of the less likely outcomes may not approach their probabilities as closely.

```python
# Generate 1000 random values from a uniform distribution
# with a minimum value of 0 and a maximum value of 1
values = np.random.uniform(0, 1, 1000)

# Plot the distribution of values
plt.hist(values, bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
